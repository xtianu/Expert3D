## 1. Reading an image:

Read the image and convert to a data matrix with pixel intensities in each color channel:

```
import skimage as sk # load skimage libraries
from skimage import io
I = sk.io.imread('/content/pawn.jpeg') # read image using imread function
```

The data matrix has the following dimensions:

```
I.shape
```

```
print('The image has {} x {} pixels and {} color channels'.format(I.shape[0],I.shape[1],I.shape[2]))
```

Visualize the image:

```
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10,6))
sk.io.imshow(I) # visualize image using imshow function
sk.io.show() 
```

Convert the RGB color image to grayscale:

```
from skimage.color import rgb2gray # conversion de RGB a escala de grises
from skimage.util import img_as_ubyte # conversion de un dato real a uint8 [0,255]

Igray = img_as_ubyte(rgb2gray(I)) # convert from RGB to grayscale
```

```
Igray.shape
```

```
Igray.dtype
```

```
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10,6))
sk.io.imshow(Igray) # visualize image using imshow function
sk.io.show() 
```

## 2. Filter an image:

Let's use a simple median filter to remove noise and smooth the image:

```
from skimage.filters import median # load the filter

from skimage.morphology import disk # load the function to define local neighborhood

sel5 = disk(25) # structuring element disk radius 5 pixels
Ifiltered = median(Igray, sel5) #Apply the median filter to obtain the filtered image

fig = plt.figure(figsize=(10,6))
sk.io.imshow(Ifiltered) # visualize image using imshow function
sk.io.show() 
```

Represent both original and filtered images side by side to compare:

```
fig = plt.figure(figsize=(14,12))
ax = fig.add_subplot(121)
ax = sk.io.imshow(Igray)
ax1 = fig.add_subplot(122)
ax1 = sk.io.imshow(Ifiltered) # visualize image using imshow function
plt.show()
```

Aplicar Filtro Gausiano

```
from skimage.filters import gaussian # load the filter

Igaussian2 = gaussian(Igray,sigma=2)
Igaussian10 = gaussian(Igray,sigma=10)

fig = plt.figure(figsize=(12,14))
ax = fig.add_subplot(131)
ax = sk.io.imshow(Igray)
plt.title('original')
ax1 = fig.add_subplot(132)
ax1 = sk.io.imshow(Igaussian2) # visualize image using imshow function
plt.title('Gaussian filter (sigma=2)')
ax2 = fig.add_subplot(133)
ax2 = sk.io.imshow(Igaussian10) # visualize image using imshow function
plt.title('Gaussian filter (sigma=10)')
plt.show()
```
