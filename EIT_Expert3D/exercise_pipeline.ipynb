```python
import matplotlib.pyplot as plt
from skimage import data
from skimage.filters import threshold_otsu

image = data.camera()
thresh = threshold_otsu(image)
binary = image > thresh

fig, axes = plt.subplots(ncols=3, figsize=(16, 6.5))
ax = axes.ravel()
ax[0] = plt.subplot(1, 3, 1)
ax[1] = plt.subplot(1, 3, 2)
ax[2] = plt.subplot(1, 3, 3, sharex=ax[0], sharey=ax[0])

ax[0].imshow(image, cmap=plt.cm.gray)
ax[0].set_title('Original')
ax[0].axis('off')

ax[1].hist(image.ravel(), bins=256)
ax[1].set_title('Histogram')
ax[1].axvline(thresh, color='r')

ax[2].imshow(binary, cmap=plt.cm.gray)
ax[2].set_title('Thresholded')
ax[2].axis('off')

plt.show()
```

# EXERCISE (PIPELINE):
1. Load an RGB color image (e.g, the astronaut or from google images)
2. Convert the RGB image to grayscale
3. Aplly a gaussian filter to smooth the image 
4. Binarize the grayscale image applying an Otsu thresholding
5. Visualize all steps in a single plot

# 1. Load RGB image

```python
from skimage import data, io, filters

I = data.astronaut()
```

# 2. convert to grayscale:

```python
from skimage.color import rgb2gray # convert RGB to Grayscale

IG = rgb2gray(I)
```

# 3. Apply a Gaussian filter

```python
from skimage.filters import gaussian # load the filter
import skimage as sk # load skimage libraries
from skimage import io

Igaussian = gaussian(IG,sigma = 5)
```

# 4. Binarize using Otsu thresholding

```python
from skimage.filters import threshold_otsu

thresh = threshold_otsu(IG)
binary = IG > thresh
```

#5. Visualize everything (original RGB, grayscale, gaussian filtered and Otsu)

```python
fig = plt.figure(figsize=(18,16))
ax = fig.add_subplot(141)
ax = sk.io.imshow(I) # original
plt.title('Original')
ax1 = fig.add_subplot(142)
ax1 = sk.io.imshow(IG) # grayscale
plt.title('Grayscale')
ax2 = fig.add_subplot(143)
ax2 = sk.io.imshow(Igaussian) # visualize image using imshow function
plt.title('Gaussian filtered')
ax3 = fig.add_subplot(144)
ax3 = sk.io.imshow(binary) # visualize image using imshow function
plt.title('Otsu binarization')
plt.show()
```
