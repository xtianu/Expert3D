Primer ejemplo de las librerías scikit image:
https://scikit-image.org/

```python
from skimage import data, io, filters # cargar los modulos que utilizaremos de las librerias skimage

image = data.coins() # carga una imagen llamada coins de la base de datos de pruebas de las librerias skimage
# ... or any other NumPy array!
edges = filters.sobel(image) #Aplicamos un filtro de Sobel de detaccion de contornos a la imagen de prueba
io.imshow(edges) # visualizamos el resultado del filtro de Sobel (imagen en la que se ven marcados los contornos)
io.show()
```

Visualizamos la imagen original coins:

```python
io.imshow(image) # visualizamos el resultado del filtro de Sobel (imagen en la que se ven marcados los contornos)
io.show()
```
